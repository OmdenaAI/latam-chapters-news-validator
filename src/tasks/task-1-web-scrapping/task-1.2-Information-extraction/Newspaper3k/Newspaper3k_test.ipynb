```python
import pandas as pd
from newspaper import Article
```

```python
df = pd.read_excel('mexico_train.xlsx')
df
```

```python
df = df.drop_duplicates(subset=['Source'])
df = df.drop(['Id', 'Category', 'Topic'], axis=1)
df
```

```python
data = []

for _, row in df.iterrows():

  d = []
  article = Article(row['Link'], language='es')
  article.download()
  try:
    article.parse()
  except:
    continue

  d.append(row['Source'])
  d.append(row['Link'])

  d.append(row['Headline'])
  d.append(article.title)

  d.append(row['Text'])
  d.append(article.text)
  
  d.append(article.authors)
  d.append(article.publish_date)

  data.append(d)
  
df_site = pd.DataFrame(data)
df_site.columns = ['Source', 'Link', 'Headline','np3k_Headline', 'article', 'np3k_article', 'np3k_Author', 'np3k_publish_date']
```
